```python
%matplotlib inline
import numpy as np
import cvxpy as cvx 
from matplotlib import pyplot as plt
```

```python
np.random.seed(0)
n = 100
k = 5
sigma = n/10

c = np.exp(-np.arange(-n/2., n/2.)**2./(2*sigma**2))/np.sqrt(2*sigma**2*np.pi)
x0 = np.zeros(n)
x0[np.random.choice(n,k)] = np.random.rand(k)*sigma
b = np.convolve(c, x0) 
b += np.random.randn(2*n-1)*np.linalg.norm(b)/np.sqrt(2*n-1)/20
plt.plot(b)
```

```python
plt.plot(np.convolve(c, x0))
```

```python
plt.plot(np.convolve(c, np.hstack((x0, np.zeros(n-1))), mode="same"))
```

```python
x = cvx.Variable(n)
f = cvx.sum_squares(cvx.conv(c, x) - b)
prob = cvx.Problem(cvx.Minimize(f), [x >= 0])
prob.solve(verbose=True)
```

```python
plt.plot(x.value)
plt.plot(x0)
```
